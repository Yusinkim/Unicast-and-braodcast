# Plot에 보내주기

```python
import socket
import time

# Using IPv4 Internet protocol and UDP protocol
server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server.bind(('', 50005))  # Server's IP address and Port Number
ip = '203.255.56.63'  # Client's IP address
port = 44440  # Client's Port Number

# The function to calculates the number of files to fragment
def frag(l):
    n = 1
    while 1:
        if l == 0:
            return 0
        if l/n >= 60000:
            n += 1
        else:
            return n

N = 1  # Number of iterations to calculate the average time

for x in range(10):
    # Reading files as a binary
    data1 = open('/Users/winlab/Desktop/A{}00.txt'.format(str(x+1)),'rb').read()
    data2 = open('/Users/winlab/Desktop/N{}00.txt'.format(str(x+1)),'rb').read()
    for k in range(N):
        start_time = time.time()  # Time to start file transfer

        l = len(data1)
        n = frag(l)

        server.sendto(str(n).encode(), (ip,port))
        time.sleep(n/1000)
        
        # First file transfer
        for i in range(n):
            server.sendto(data1[i*l//n : (i+1)*l//n], (ip, port))
            time.sleep(0.001)
            
        time.sleep(1)
        
        # Second file transfer
        for i in range(n):
            server.sendto(data2[i*l//n : (i+1)*l//n], (ip, port))
            time.sleep(0.001)

        server.sendto(str(start_time).encode(), (ip,port))
        time.sleep(1)

server.close()
```
