```python
import numpy as np
import socket
import time

# The function to XOR two files
def XOR(file1, file2):
    data1 = open(file1, 'rb').read()
    data2 = open(file2, 'rb').read()
    
    data1 = np.frombuffer(data1, dtype='uint8')
    data2 = np.frombuffer(data2, dtype='uint8')

    xor_data = np.bitwise_xor(data1, data2)
    xor_data = xor_data.tostring()
    
    return xor_data

# Using IPv4 Internet protocol and UDP protocol
server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server.bind(('', 50000))  # Server's IP address and Port Number
ip = '203.255.56.63'  # Client's IP address
port = 44444  # Client's Port Number

# The function to calculates the number of files to fragment
def frag(l):
    n = 1
    while 1:
        if l == 0:
            return 0
        if l/n >= 60000:
            n += 1
        else:
            return n

N = 1  # Number of iterations to calculate the average time

for x in range(10):
    # The path and name of files
    file1 = '/Users/winlab/Desktop/A{}00.txt'.format(str(x+1))
    file2 = '/Users/winlab/Desktop/N{}00.txt'.format(str(x+1))
    for k in range(N):
        start_time = time.time()  # Time to start file transfer

        data = XOR(file1, file2)  # XOR operation

        l = len(data)
        n = frag(l)

        server.sendto(str(n).encode(), (ip, port))
        time.sleep(n/2000)
        
        # File transfer
        for i in range(n):
            server.sendto(data[i*l//n : (i+1)*l//n], (ip ,port))
            time.sleep(0.001)

        server.sendto(str(start_time).encode(), (ip ,port))
        time.sleep(1)

server.close()
```
